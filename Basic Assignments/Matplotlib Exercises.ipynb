** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig = plt.figure( figsize = (4,2))


axes = fig.add_axes([0,0,1,1])
axes1 = fig.add_axes([0.1,0.5,.4,.4])

axes.plot(x,z,'b')
axes1.plot(x,y,'b')
axes1.set_xlim([20,22])
axes1.set_ylim([30,50])
axes1.set_xlabel('X')
axes1.set_ylabel('Y')


# axes.plot(x,y, color = 'blue', alpha =0.6)
fig
```
