### Assignment 6: Time Series Analysis

1. Create a Pandas DataFrame with a datetime index and one column filled with random integers. Resample the DataFrame to compute the monthly mean of the values.

```python
date_rng = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')
df13 = pd.DataFrame(date_rng, columns=['date'])
df13['data'] = np.random.randint(0, 100, size=(len(date_rng)))

df13['month'] = df13['date'].dt.to_period('M')
df13.groupby(['month'], as_index=False)['data'].mean()
```

2. Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-12-31' and one column filled with random integers. Compute the rolling mean with a window of 7 days.

```python
date_rng = pd.date_range(start='2021-01-01', end='2021-12-31', freq='D')
df14 = pd.DataFrame(date_rng, columns=['date'])
df14['data'] = np.random.randint(0, 100, size=(len(date_rng)))


df14['rolling_mean'] = round( df14['data'].rolling(window= 7).mean(),1)
df14['rolling_mean_forward'] = round( df14['data'].rolling(window= 7).mean().shift(-6),1)
df14
```

```python
df14['month'] = df14['date'].dt.to_period('M')
df15 = df14.groupby(['month'],as_index=False).agg({'data': [np.sum, np.mean]})

df15.columns = ["_".join(col) for col in df15.columns]

df15['rolling_mean'] = round( df15['data_sum'].rolling(window= 3).mean(),1)
df15['rolling_mean_forward'] = df15['data_sum'].rolling(3).mean().shift(-(3-1)) ##forward 2 shifted back
df15
```

### Assignment 8: Pivot Tables

1. Create a Pandas DataFrame with columns 'Date', 'Category', and 'Value'. Create a pivot table to compute the sum of 'Value' for each 'Category' by 'Date'.

```python
date_rng = pd.date_range(start='2022-01-01', end='2022-01-10', freq='D')
df17 = pd.DataFrame({'Date': np.random.choice(date_rng, size=20), 'Category': np.random.choice(['A', 'B', 'C'], size=20), 'Value': np.random.randint(1, 100, size=20)})

df17.groupby(['Category','Date'], as_index=False)['Value'].sum().pivot(columns='Category',values= 'Value', index = 'Date')
```

```python
df17.pivot_table(columns='Category',values= 'Value', index = 'Date',aggfunc=sum)
```


2. Create a Pandas DataFrame with columns 'Year', 'Quarter', and 'Revenue'. Create a pivot table to compute the mean 'Revenue' for each 'Quarter' by 'Year'.
