```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Enable inline plotting
%matplotlib inline

# Example dataset
tps = sns.load_dataset("tips")

tps.head()
```

```python
# Show the histogram
##default hist
g = sns.displot(tps['total_bill'],kde = True, bins = 50)
#kde = is for line, bins = how many bars in chart
g.figure
```

```python
#line
g = sns.displot(tps['total_bill'], kind = 'kde')
g.figure
```

```python
##cdf
g = sns.displot(tps['total_bill'], kind = 'ecdf')
g.figure
```

```python
g = sns.displot(x = tps['tip'], y = tps['total_bill'])
#kde = is for line, bins = how many bars in chart
g.figure
```

```python
gi = sns.scatterplot(x = tps['tip'], y = tps['total_bill'],)
# displot(x = tps['tip'], y = tps['total_bill'])
#kde = is for line, bins = how many bars in chart
gi.figure
```

```python
gi = sns.jointplot(x = tps['tip'], y = tps['total_bill'])
gi.figure
```

```python

gi = sns.jointplot(x = tps['tip'], y = tps['total_bill'], kind= 'hex')
gi.figure
```

```python
giq = sns.pairplot(tps)
giq.figure
```

```python
giq = sns.pairplot(tps, hue = 'sex')
giq.figure
```

# Categorical Plots

```python
import numpy as np
```

```python
#by default it plot mean
plt.clf()  # Clear the current figure (optional but safe)
plt.figure() 
```

```python
g_bar = sns.barplot(data = tps , x = 'sex', y = 'total_bill')
g_bar.figure
```

```python
#by default it plot mean
plt.clf()  # Clear the current figure (optional but safe)
plt.figure() 
```

```python
g_bar = sns.barplot(data = tps , x = 'sex', y = 'total_bill',estimator = np.std)
g_bar.figure
```

```python
plt.clf()  # Clear the current figure (optional but safe)
plt.figure()

g_bar = sns.countplot(data = tps , x = 'sex',color = 'r')
g_bar.figure
```

```python
plt.clf()  # Clear the current figure (optional but safe)
plt.figure()

g_bar = sns.boxplot(data = tps , x = 'day', y = 'total_bill',color = 'r')
g_bar.figure
```

```python
plt.clf()  # Clear the current figure (optional but safe)
plt.figure()

##hue allows to color category
g_bar = sns.boxplot(data = tps , x = 'day', y = 'total_bill', hue = 'smoker')
g_bar.figure
```

```python
plt.clf()  # Clear the current figure (optional but safe)
plt.figure()

##hue allows to color category
g_bar = sns.violinplot(data = tps , x = 'day', y = 'total_bill', hue = 'smoker',split = True)
g_bar.figure
```

```python
plt.clf()  # Clear the current figure (optional but safe)
plt.figure()

##hue allows to color category
g_bar = sns.stripplot(data = tps , x = 'day', y = 'total_bill', hue = 'smoker',jitter =True, dodge = True)
g_bar.figure
```

```python
plt.clf()  # Clear the current figure (optional but safe)
plt.figure()

##hue allows to color category
g_bar = sns.swarmplot(data = tps , x = 'day', y = 'total_bill')
g_bar.figure
g_bar1 = sns.violinplot(data = tps , x = 'day', y = 'total_bill')
g_bar1.figure
```

```python
plt.clf()  # Clear the current figure (optional but safe)
plt.figure()

##hue allows to color category
# g_bar = sns.swarmplot(data = tps , x = 'day', y = 'total_bill')
# g_bar.figure
g_bar1 = sns.catplot(data = tps , x = 'day', y = 'total_bill', kind = 'bar' )
g_bar1.figure
```

# Heatmaps

```python
fl = sns.load_dataset("flights")
```

```python
fl.head()
```

```python
fli = fl.pivot_table(index = 'month', columns = 'year',values = 'passengers',observed=True)
```

```python
fli.head()
```

```python
plt.clf()  # Clear the current figure (optional but safe)
plt.figure()

ht1 = sns.heatmap(data = fli, cmap = "coolwarm", linewidths= 1)
ht1.figure
```

```python
plt.clf()  # Clear the current figure (optional but safe)
plt.figure()

ht1 = sns.clustermap(data = fli, cmap = "coolwarm", linewidths= 1)
ht1.figure
```
